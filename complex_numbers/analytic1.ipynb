This analysis focuses on the scenario where we have only one intermittently laid cable.  This model can assume the characteristics of the double-cable configuration because the signal from the non-intermittent cable is effectively just an additive component to the background.

We initially just measure the baseline.  
$$m_1 = b$$

From the diagram we add up phasors to arrive that the measured complex signal when the cable is present.
$$m_2 = b + \Delta b + s$$

Taking the difference of the two measurments, a couple steps of algebra arrive at a definition
for the complex number $z$.  The complex angle of $z$ is simply the angle between the cable and measurement.
$$m_2 - m_1 = s + \Delta b$$
$$z \equiv \frac{s}{m_2} = 1 - \frac{m_1 + \Delta b}{m_2}$$
$$z = 1 - \frac{b + \Delta b}{b + \Delta b + s}$$

Expressing $z$ as a function of $\Delta\left(b\right)$ allows us to take the ratio of complex numbers.  The complex angle of this ratio will be the angle variation introduced by adding the changing background $\Delta b$.

$$z\left(\Delta b\right) = \frac{\frac{s}{b}}{1 + \frac{\Delta b}{b} + \frac{s}{b}}$$
$$z_0 \equiv z\left(0\right) = \frac{\frac{s}{b}}{1 + \frac{s}{b}}$$
$$\frac{z}{z_0} = \frac{1 + \frac{s}{b}}{1 + \frac{s}{b} + \frac{\Delta b}{b}}$$

So lets take the strategy:
* Choose $\left|\frac{s}{b}\right|$ and $\left|\frac{\Delta b}{b}\right|$
* Choose random angle over $2 \pi$ for each of those fractions to transform to complex numbers.
  This basically models the idea that the background can take on any angle with respect to the signal
  and that the change in background can take on any angle with respect to the background.  The two
  angles are independentaly random.
* For each choice of random variables the complex angle $\theta = arg\left(\frac{z}{z_0}\right)$ is computed.
* The fraction of these combinations that have $\theta$ less than some threshold is a proxy for the probability
  of that combination being "good" or "bad".
  
  
  
Also... random fact.  Look up "von Mises distribution" in the scenario you need to model distributions over angles.
