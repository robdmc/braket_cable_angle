```python

def get_las_objects(p):
    
    out = {'samples': [], 'full_samples': []}
    
    las_back_all = Las(
        p.back_file, positive_angles=False, name=os.path.basename(p.back_file), normalize=p.normalize)
    
    out['back_all'] = las_back_all

    # las_back = las_back_all.section('background', min_time=1260, max_time=1320)
    las_back = las_back_all.section('Background', min_time=p.back_times[0], max_time=p.back_times[1])
    out['back'] = las_back
    
    for file_name, time_range, shift in zip(p.sample_file_names, p.sample_ranges, p.shifts):
        las_full = Las(file_name, file_name, normalize=normalize)
        las = las_full.section(
            os.path.basename(file_name), min_time=time_range[0], max_time=time_range[1],
            shift=shift
        )
        out['samples'].append(las)
        out['full_samples'].append(las_full)
        
        
    return out
```

```python
def normalize(df, shift=0):
    df = df.sort_values(by='bearing')
    y = df.signal
    y = y - y.mean()
    y = y / y.std()
    y = np.roll(y, shift)
    df.loc[:, 'signal'] = y
    
    return df
```

```python
def adjust_lims(ax):
    y_min, y_max = ax.get_ylim()
    new_min = y_min - 0.1 * (y_max - y_min)
    new_max = y_max + 0.2 * (y_max - y_min)
    ax.set_ylim((new_min, new_max))
    return ax
    
def time_plot(las, bounds_las=None):
    title = os.path.basename(las.name)
    
    df = las.df
    ax = ezr.figure(style='seaborn-darkgrid')
    ax.plot(df.time, df.signal, label='Signal')
    ax.plot(df.time, df.bearing, label='Bearing')
    ax.set_title(title)
    
    if bounds_las is not None:
        y_min, y_max = ax.get_ylim()
        t_back_min, t_back_max = bounds_las.df.time.min(), bounds_las.df.time.max()

        ax.fill_between(
            [t_back_min, t_back_max], [y_max, y_max], [y_min, y_min],
            alpha=.2,
            color=ezr.cc.d,
            label='Selected'
        )


    ax.legend(loc='upper center', ncol=3)
    ax.set_xlabel('Time (s)')
    ax.set_yticklabels([])
    
    ax = adjust_lims(ax)
    
    file_name = os.path.splitext(title)[0] + '.png'
    pl.figure(ax.figure.number)
    pl.savefig(file_name)
    

def bearing_plot(back_las, sample_las_list, title=None):
    ax = ezr.figure(style='seaborn-darkgrid')
    df = normalize(back_las.df)
    ax.plot(df.bearing, df.signal, 'k', label='Background')
    for las in sample_las_list:
        df = normalize(las.df, las.shift)
        ax.plot(df.bearing, df.signal, label=las.name, linewidth=6, alpha=.5)
    ax.legend(loc='upper center', ncol=3)
    ax.set_xlim((-180, 180))
    ax.set_xlabel('Bearing (deg)')
    ax.set_ylabel('Normalized Signal')
    ax = adjust_lims(ax)
    if title is not None:
        ax.set_title(title)
        
    file_name = title + '.png'
    pl.figure(ax.figure.number)
    pl.savefig(file_name)
#     ax.set_ymargin(1.4)
    
```

```python
def make_plots(p):
    res = get_las_objects(p)
    time_plot(res['back_all'], res['back'])

    for full_sample, sample in zip(res['full_samples'], res['samples']):
        time_plot(full_sample, sample)

    bearing_plot(res['back'], res['samples'], title=p.data_set)    
    
```
